```julia
#ejercicio 1
println("Ingresa un número entero h, 1 ≤ h ≤ 12\n")
h = parse(Int64,readline())
println("Ingresa un número entero m, 0 ≤ m ≤ 59\n")
m = parse(Int64,readline())
Números = []
#Regla de tres
ang2 = m*6
ang1 = 30*(h + (ang2/360))
if ang1 >= 360
    ang1 = 30*(ang2/360)
end
append!(Números,ang1)
append!(Números,ang2)
print("Ángulo de la manecilla de:\nhoras, minutos\n")
print(Números)
```

```julia
#ejercicio 2
using Plots
fig = Plots.plot()
x = range(((-3*pi)/2),stop=((2*pi/2)),length=100)
y = [sin(i) for i in x]
y2 = [cos(i) for i in x]
y3 = [tan(i) for i in x]
Plots.plot!(x,y,label = "sin(x)",ylims=(-3,3))
Plots.plot!(x,y2,label = "cos(x)",ylims=(-3,3))
Plots.plot!(x,y3,label = "tan(x)",ylims=(-3,3))
Plots.title!("Ejercicio 2",fontsize=24)
display(fig)
```

```julia
#ejercicio 6

#serie
function mi_serie(k,x)
    suma = 0
    while k >= 1
        suma = suma + (4/pi)*(1/(2*k-1))*sin(2*pi*x*(2*k - 1))
        k = k - 1
    end
    return suma
end

#función -1,1
function f(x)
    if x < 0
        return -1
    else
        return 1
    end
end
```

```julia
#ejercicio 7    
fig = Plots.plot()
x = range(-1/2,stop=1/2,length=1000)    

y = [mi_serie(1,i) for i in x]
y2 = [mi_serie(3,i) for i in x]
y3 = [mi_serie(5,i) for i in x]
y4 = [mi_serie(7,i) for i in x]
y5 = [f(i) for i in x]

Plots.plot!(x,y,label="mi_serie(1,x)",ylims=(-1.5,1.5))
Plots.plot!(x,y2,label="mi_serie(3,x)")
Plots.plot!(x,y3,label="mi_serie(5,x)")
Plots.plot!(x,y4,label="mi_serie(7,x)")
Plots.plot!(x,y5,label="f(x)",legend=:topleft)
Plots.title!("Ejercicio 7", frontsize=24)
#Plots.savefig("prueba.pdf")
display(fig)
```
