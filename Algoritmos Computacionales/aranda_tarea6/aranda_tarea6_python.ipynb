```python
#ejercicio 1
h = int(input("Ingresa un número entero h, 1 ≤ h ≤ 12\n"))
m = int(input("Ingresa un número entero m, 0 ≤ m ≤ 59\n"))
Números = []
#Regla de tres
ang2 = m*6
ang1 = 30*(h + (ang2/360))
if ang1 >= 360:
    ang1 = 30*(ang2/360)
Números.append(ang1)
Números.append(ang2)
print("Ángulo de la manecilla de:\n{0}, {1}".format("horas", "minutos"))
print(Números)
```

```python
#ejercicio 2
import numpy as np
import matplotlib.pyplot as plt
a = -3*np.pi/2
b = 3*np.pi/2
n = 1000
x = np.linspace(a,b,n)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.ylim(-3,3)
plt.plot(x,y,x,y2,x,y3)
plt.legend(("sin(x)","cos(x)","tan(x)"),loc = "best")
```

```python
#ejercicio 6
def mi_serie(k,x):
    suma = 0
    while k >= 1:
        suma = suma + (4/np.pi)*(1/(2*k-1))*np.sin(2*np.pi*x*(2*k - 1))
        k = k - 1
    return suma


def f(x):
    if x < 0:
        return -1
    else:
        return 1
```

```python
#ejercicio 7
a, b, n = -1/2, 1/2, 1000
x = np.linspace(a,b,n)
Yf = []
for i in x:
    Yf.append(f(i))
y = mi_serie(1,x)
y2 = mi_serie(3,x)
y3 = mi_serie(5,x)
y4 = mi_serie(7,x)
plt.plot(x,y,x,y2,x,y3,x,y4,x,Yf,"k")
plt.legend(("mi_serie(1,x)","mi_serie(3,x)","mi_serie(5,x)","mi_serie(7,x)","f(x)"),loc="best")
```
