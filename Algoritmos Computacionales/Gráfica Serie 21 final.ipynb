```python
#ejercicio 6
import numpy as np
import matplotlib.pyplot as plt

def mi_serie(n,x,t):
    suma = 0
    while n >= 1:
        suma = suma + np.exp(-t)*np.sin(n*x)*(2/np.pi*n*(n-1)**2)*((n*(t-1)-t)+n) + t*np.exp(-t)*(1- x/np.pi)        
        n = n - 1
    return suma
```

```python
#ejercicio 7
a, b, n = 0, np.pi, 1000
x = np.linspace(a,b,n)

y = mi_serie(100,x,100)
plt.plot(x,y)
plt.legend(("solución",),loc="best")
```
